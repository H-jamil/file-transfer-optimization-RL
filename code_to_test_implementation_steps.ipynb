```python
import multiprocessing as mp
import time
import numpy as np
from config import configurations
import os
class TestClass:
    def __init__(self,configurations):
        self.num_workers=4
        self.file_count=4
        self.root=configurations["data_dir"]
        self.process_status = mp.Array("i", [0 for i in range(self.num_workers)])
        self.file_names = os.listdir(self.root) * configurations["multiplier"]
        manager=mp.Manager()
        self.q = manager.Queue(maxsize=self.file_count)
        self.th_list=manager.list()
        for i in range(self.file_count):
            self.q.put(i)
#         self.file_sizes = [os.path.getsize(self.root+filename) for filename in self.file_names]
        self.file_sizes = mp.Array("i",[os.path.getsize(self.root+filename) for filename in self.file_names])
        for i in self.file_sizes:
            print(f"{i}")

    def worker(self,id,q):
        k=0
        while True:
            if self.process_status[id]==0:
                self.process_status[id]=1
            else:
                self.process_status[id]=0
            q.get()
            print(f"Process {id} says {q} and value is {q.qsize()}\n")
            time.sleep(0.5)
            if q.qsize()>=0:
                q.put(id)
                print(f"Process {id} says {q} is empty and it puts to make it {q.qsize()}\n")
            k+=1
            if k==10:
                break
            
    
    def monitor(self):
        count=0
        while True:
            print(f"#################")
            print(f"From monitor Thread total active process are {self.q} and {self.q.qsize()} ********")
            time.sleep(1)
            count+=1
            if count ==10:
                break
    
    def run(self):
        workers = [mp.Process(target=self.worker, args=(i,self.q)) for i in range(self.num_workers)]
        for p in workers:
            p.daemon = True
            p.start()
        reporting_process = mp.Process(target=self.monitor)
        reporting_process.daemon = True
        reporting_process.start()
        return workers,reporting_process


class TestTestClass:
    def __init__(self,testClass):
        self.testClass=testClass
```

```python
testObject=TestClass(configurations)
testTestObject=TestTestClass(testObject)
print(f"Value from main thread is {testObject.q} and {testObject.q.qsize()}")
print(f"Value from main main thread is {testTestObject.testClass.q} and {testTestObject.testClass.q.qsize()}")
testObject.run()
```
