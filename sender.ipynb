```python
import os
import math
import time
import socket
import datetime
import numpy as np
import logging as log
import multiprocessing as mp
import re
root="/home/hjamil/Documents/files_bash/"
file_names="FILE0"
chunk_size=1024*1024
offset=0
file_size=os.path.getsize(root+file_names)
to_send = file_size - offset
sock = socket.socket()
sock.settimeout(3)
sock.connect(("128.205.223.105",50023))
print("socket connected")
filename = root + file_names
msg = filename + "," + str(int(offset))
msg += "," + str(int(to_send)) + "\n"
try:
    sock.send(msg.encode())
    print("starting {0}, {1}".format(file_names, filename))
except:
    print("Process {process_id} failed to send file message")

while (to_send>0):
    block_size = int(min(chunk_size,to_send))
    data_to_send = bytearray(int(block_size))
    try:
        sent = sock.send(data_to_send)
    except:
        sent=0
        print("socket error")
    offset += sent
    to_send -= sent
    print("block sent")
sock.close()
```
